One key feature of fully developed turbulence is the wide range of scales due to the high Reynolds number. The scales can be loosely classified as large scales and small scales. Large scales contain most of the turbulent kinetic energy and dominant in momentum, mass and heat transport, whereas small scales are mainly responsible for energy dissipation. These classifications are mainly based on the 1941 Kolmogrov theory which is stated in the form of three hypotheses, i.e. Kolmogrov’s hypothesis of local isotropy, first similarity hypotheses and second similarity hypotheses [1]. Stemming from these hypotheses, the Kolmogrov scale is
described in the form of viscosity $\nu$ and energy dissipation rate $\epsilon$, and the turbulence spectrum is divided into three different ranges, i.e. the energy-containing range, the inertial range, and the dissipation range. For large enough Reynolds number, the energy-containing range and the dissipation range should be wide separated in the wavenumber space, as you shall see in this excersie. Four characteristic wave numbers are important here. $k_l$ and $k_\eta$ should mark the maximum of the energy spectrum and the dissipation rate spectrum, respectively. $K_l$
and $K_\eta$ are the starting point and end point of the inertial range, in which both the energy and dissipation rate should be small compared to the total values. The relationship between these wavenumbers is $k_l$ < $K_l$ < $K_\eta$ < $k_\eta$. 

The 1941 Kolmogrov hypotheses are fundamental to many of the modern turbulence theories. Enormous efforts have been put on experimental tests as well as direct numerical simulations to verify and advance the theories. A major limitation for these efforts is that it is difficult to generate large enough Reynolds number to create the separation between the energy and dissipation rate spectrum. Here, we are introducing the so-called model spectrum that incorporates key observations from Komogorov's 1941 hypothesis. 

### kinematic representation of the velocity using Fourier series

To better understand the relationship between different turbulent scales, the behavior of the turbulent spectrum and the velocity statistics in a flow field, a kinematic fluctuating velocity is constructed over a domain of length $L=2\pi$ using Fourier series as shown in Eq. (1).

$ u(x) = \sum^{N}_{n=1} a_n \sin (k_n x + \psi_n) $.  (1)

Each term in Eq. (1) is called a "Fourier mode", with $a_n$ being the "Fourier coefficient" and $\psi_n$ the phase angle for the particular Fourier mode. The reason that Eq. (1) is a useful representation for turbulence is that it is a series in which each term represents a contribution with a well-defined length scale $\lambda_n$ where $\lambda_n = 2\pi/k_n$. When we talk about a particular length scale $r$ in turbulence, we are mathematically talking about a group of Fourier modes with contiguous scales $\lambda_n$ over a selected range of $k_n$. The particular length scale $r$ can be defined as a mean of the $\lambda_n$ in the chosen range of $k_n$. 

When a domain length $L=2\pi$ is chosen, the wavenumber $k_n = 2\pi n/L = n$. The phase angle $\psi_n$ can be treated as a random number ranging between 0 and $2\pi$. The amplitude $a_n$ is related to the energy contained by each mode. To make the velocity representative of a true isotropic homogeneous turbulence field, a modeled turbulent kinetic energy spectrum is employed to assign values for $a_n$, as discussed next. 

### Modeled spectrum 

A variety of turbulence model spectra exists, for example, the von Karman spectrum introduced in Eq.(2) or the model spectrum introduced in (S. Pope, Turbulent flows, page 232). 

$ E_{VK} (k_n) = C \frac{\sigma^2}{k_e} \frac{(k_n/k_e)^4}{[1+(k_n/k_e)^2]^{17/6}} $,     (2)

where $C = \frac{55}{9} \frac{\Gamma(5/6)}{\sqrt(\pi)\Gamma(1/3)}$, $\sigma^2$ is the variance of the outer-scale velocity.


When $(k_n/k_e)$ is large enough, Eq. (2) is asymptotic to the -5/3 power relationship between $E_{VK}$ and $k_n$. You might have hear the -5/3 relation as it is a popular way to verify whether the turbulence generated in a simulation is realistic or not. For turbulence generated in 2D, the power law relation usually has a -3 or -4 power relation (https://doi.org/10.1016/0021-9991(77)90023-7), which is why 2D turbulence does not represent realistic features of 3D turbulence.

When $k_n$ is large enough, the spectrum should fall into the dissipation range in which $E_{VK}$ decays much faster than in the inertial range. To ensure the dissipation range is correctly represented, the von Karman spectrum $E_{VK}$ is multiplied by an additional term to accelerate its dissipation, 

$E(k_n) = E_{VK} \exp(-(k_n - K_\eta)^2/\sigma^2_k)$
where $\Gamma$ is the Gamma function, and $\sigma_k$ is the wave number that is closely related to the Kolmogorov scale $\eta$. 

Finally, the model spectrum is related to the Fourier series of the fluctuating velocity in Eq. (1) as, 

$E(k_n) = a^2_n$

Once $E(k_n)$ is determined from the model spectrum, one can reconstruct the fluctuating velocity in physical space through the Fourier series. 

### Parameterization for the modeled spectrum


To further assign values to the constants in Eq. (2), we first introduce the relations between a few commonly used Reynolds numbers. An outer scale (or integral scale) Reynolds number is usually defined using the integral length scale ($l$, which corresponds to the wavenumber $k_l$) as 

$ Re_l = \frac{\sigma l}{\nu} $,
where $\sigma$ is a velocity scale that is defined based on the total turbulent kinetic energy. 
The outer scale Reynolds number represents the degree of the scale separation, i.e., the ratio of the largest to smallest length scales in turbulence, i.e., 

$ Re_l = (\frac{l}{\eta})^{(4/3)} $.

We further define $R_I = \frac{K_\eta}{K_l}$, which characterizes the length of the so-called inertial range. Therefore, both $Re_l$ and $Re_I$ represent the scale separation in turbulence, and they are closely related. Another relevant Reynolds number is the Taylor micro-scale Reynolds number. The Taylor micro-scale (conventionally denoted as $\lambda$) represents the length scale in the inertial range, and is related to the outer scale (integral scale) Reynolds number $Re_l$ to be 

$Re_\lambda = \frac{\sigma \lambda}{\nu}= \sqrt(Re_l)$

In the turbulence literature, $Re_\lambda$ is a popular choice in terms of reporting the Reynolds number. 

To actually assign numbers to the model spectrum, Let's choose $k_e$ to be 5 in this exercise. The peak value of the spectrum is defined to correspond to the integral scale, i.e., $k_l = 1.549 k_e = 7.75$ (can be figured out by taking derivative of Eq. (2)). The start of the inertial scale $K_l$ is chosen to be $10k_l = 77$, and the end of the inertial range $K_\eta$ can be expressed using $R_I$ as $K_\eta = R_I K_l$. $k_\eta$ is chosen to be 70 $K_\eta$. The integral velocity $\sigma$ is related to the Reynolds number $Re_l$ as $\sigma = \nu Re_l/l$. The wavenumber $\sigma_k$ is specified to be $\sigma_k = k_\eta/10$. 

The spectrum can be split into 3 ranges of scales for analysis below: 

I. large eddies, or integral scales, $k \approx 1 - K_l$ 

II. inertial range, $k \approx K_l - K_\eta$

III. dissipation range, $k \approx K_\eta - k_{max}$

We will choose $k_{max} = 2k_\eta$. 

The subsequent code implements the turbulent kinetic energy spectrum and the dissipation rate spectrum. The velocity is constructed through the Fourier's series described above. You are expected to further explore the concept of large-scale and small-scale phenomena through guided plotting exercise.

```python
import numpy as np
import math
import time
import matplotlib.pyplot as plt

PI = math.pi
## Build a spectrum for RI = K_\eta/K_l = 10, which corresponds to Re_l = 133905.2 and Re_\lambda = 365.9.
def build_spectrum(
    Kmax: int,
    ke: float = 5.0,
    sigma: float = 2.6,
    nu: float = 1.5e-5,
    Kyita: int = 775,
    sigmak: float = 5245.0,
    seed: int | None = None,
    urms_target: float | None = None,
):
    """
    Returns k (1..Kmax), E(k), D(k)=2*nu*k^2*E(k), amplitudes a_k, phases phi_k.
    Physically consistent mapping: u(x)=Σ a_k sin(kx+phi_k) with a_k = sqrt(2 E_k).
    """
    rng = np.random.default_rng(seed if seed is not None else int(time.time()))
    k = np.arange(1, Kmax + 1, dtype=np.float64)

    # Physically clean: floating division 6/5
    C = 55.0 * (6.0 / 5.0) * 0.940664 / 9.0 / math.sqrt(PI) / 3.0 / 0.892994

    # Your original functional form for the spectrum "shape"
    E = (C * sigma * sigma * (k / ke) ** 4 / ke) / (1.0 + (k / ke) ** 2) ** (17.0 / 6.0)

    # Dissipation-range cutoff for k > Keta (Gaussian)
    if Kyita < Kmax:
        kk = k[Kyita:]  # k = Kyita+1 .. Kmax
        cutoff = np.exp(-((kk - Kyita) / sigmak) ** 2)
        E[Kyita:] *= cutoff

    # Amplitudes for synthesized field
    a = np.sqrt(E)
    #a = np.sqrt(E)
    # Phase angle of the velocity field
    phi = 2.0 * PI * rng.random(Kmax)
    # Dissipation rate is related to kinetic energy mode by the following in isotropic turbulence
    D = 2.0 * nu * (k ** 2) * E

    return k, E, D, a, phi


def plot_spectrum_with_kolmogorov_ref(k, E, D, Kl: int, Kyita: int):
    """
    Plots E(k) and D(k), and overlays a -5/3 reference slope in the inertial range [Kl, Kyita].
    """
    Kmax = int(k[-1])
    k1 = max(1, Kl)
    k2 = min(Kmax, Kyita)

    mask = (k >= k1) & (k <= k2)
    k_inert = k[mask]

    if len(k_inert) < 2:
        raise ValueError("Inertial-range window is empty or too small for the current Kmax.")

    # Anchor the reference line at geometric-mean k0 for a fair log-space comparison
    k0 = math.sqrt(k1 * k2)
    idx0 = int(np.argmin(np.abs(k - k0)))
    E0 = float(E[idx0])

    ref = E0 * (k_inert / k0) ** (-5.0 / 3.0)

    # Plot E(k) with -5/3 line
    Enorm = E/max(E)/sum(E)
    Dnorm = D/max(D)/sum(D)
    plt.figure()
    plt.loglog(k, Enorm, label="E(k)")
    plt.loglog(k_inert, ref, label=r"Reference slope $k^{-5/3}$")
    plt.loglog(k, Dnorm, label=r"D(k) = 2 ν k² E(k)")
    plt.axvspan(k1, k2, alpha=0.15, label="Inertial range window")
    plt.xlabel("k")
    plt.ylabel("E(k) or D(k)")
    plt.title(r"Turbulent kinetic energy and dissipation spectrum")
    plt.grid(True, which="both")
    plt.ylim(1E-4, 1)
    plt.legend()


    plt.show()


def main():
    # ---- match your original parameters ----
    ke = 5.0
    Kl = 77
    Kyita = 775
    kyita = 70*Kyita
    Kmax = 2*kyita

    # Build spectrum (set urms_target to control energy level, e.g., 1.0)
    k, E, D, a, phi = build_spectrum(
        Kmax=Kmax,
        ke=ke,
        #Kl=Kl,          # not used inside build_spectrum, kept for clarity
        Kyita=Kyita,
        urms_target=None
    )
    # Plot the turbulent kinetic energy spectrum and the dissipation rate spectra togeter
    plot_spectrum_with_kolmogorov_ref(k, E, D, Kl=Kl, Kyita=Kyita)

    # Using the output a and phi, you can look at the velocity statistics described below.

if __name__ == "__main__":
    main()
```

The sum of E(k) is essentially the turbulent kinetic energy and the sum of D(k) is the dissipation rate. The famous two-equation RANS k-epsilon model essentially models $\sum E(k)$ and $ \sum D(k)$ examined here. Since the sum of these fourier modes are considered, RANS models do not contain any scale information (That's why they appear to be so smooth even though they are modeling turbulent flows). 

The above clearly shows the scale separation between the large energy-containing scale and the small energy-dissipation scale. This is because the integral Reynolds number (or the Taylor Reynolds number) are relatively large in this example. Conventionally, $Re_\lambda = 100$ is a very large Reynolds number simulation in the world of direct numerical simulation. This is also why affordable DNS at smaller Reynolds number often suffers from lack of inertial range in their data. A large DNS database on turbulent flow can be found on https://turbulence.idies.jhu.edu/home Interested readers can explore the data on the webiste and plot their kinetic energy spectra. 

You can try increasing $R_I$ or decreasing $R_I$ to observe the relative positions of the two spectrum and the traits of the resulting velocity field, if interested. 

Largely, for large eddy simulation (LES), a filter $\Delta$ is usually selected in the inertial range, often near $K_\eta$. The scales where $k < K_\eta$ are resolved by the mesh grid, and any scale information with a wavenumber larger than $\Delta$ requires modeling. That's why LES does involve scale information, although the scale-scale information is lost. The so-called well-resolved LES usually indicates that 90\% of the turbulent kinetic energy is resolved by the grid, which implies that the filter is located near $K_\eta$ towards $k_\eta$.

```python
# parameters from original example
ke = 5.0
Kl = 77
Kyita = 775
kyita = 70*Kyita
Kmax = 2*kyita

# build model spectrum
k_vals, E_vals, D_vals, a_vals, phi_vals = build_spectrum(
    Kmax=Kmax, ke=ke, Kyita=Kyita, seed=42 
)

# define mesh
Nx = 5 * Kmax
x = np.linspace(0, 2 * PI, Nx, endpoint=False)

print(f"Kmax = {Kmax:,} | Nx = {Nx:,} | K_eta = {Kyita} | k_eta = {kyita}")
```

```python
def reconstruct_range(k_low, k_high):
    # ----------------------------------------------------------------------------------
    # THE COMMENTED CODE BELOW WORKS BUT IS VERY SLOW (runtime = 15 minutes 30 seconds)
    # ----------------------------------------------------------------------------------

    # idx_s = int(k_low) - 1
    # idx_e = int(k_high)

    # k_slice = k_vals[idx_s:idx_e]
    # a_slice = a_vals[idx_s:idx_e]
    # phi_slice = phi_vals[idx_s:idx_e]

    # u_sum = np.zeros(Nx)

    # for i in range(len(k_slice)):
    #     u_sum += a_slice[i] * np.sin(k_slice[i] * x + phi_slice[i])
    # return u_sum

    # instead we can use numpy's Inverse Fast Fourier Transform (IFFT), runtime = 0.4 seconds
    C = np.zeros(Nx, dtype=complex)
    idx = np.arange(k_low, k_high + 1)

    # re-write sin as a complex exponential 
    C[idx] = a_vals[idx -1] * np.exp(1j * phi_vals[idx - 1])

    # compute the sum of complex exponentials using IFFT at every grid point,
    # multiply by Nx do undo IFFT scaling, take imaginary part corresponding to the original sin function 
    return np.imag(Nx * np.fft.ifft(C))

u_integral       = reconstruct_range(1, Kl)                 # Range I:   k = 1 ... 77
u_inertial       = reconstruct_range(Kl + 1, Kyita)         # Range II:  k = 78 ... 775
u_dissipation    = reconstruct_range(Kyita + 1, Kmax)       # Range III: k = 776 ... 108 500

fig, axs = plt.subplots(3, 1, figsize=(12, 9))

# Range I
axs[0].plot(x, u_integral, color='tab:blue', lw=0.8)
axs[0].set_title(fr"Range I: Energy-Containing (Integral): $1 \leq k \leq {Kl}$")
axs[0].set_xlabel("x [rad]")
axs[0].set_ylabel("u [m/s]")

# Range II
m1 = x < 0.5
axs[1].plot(x[m1], u_inertial[m1], color='tab:orange', lw=0.8)
axs[1].set_title(fr"Range II: Inertial Range: ${Kl} \leq k \leq {Kyita}$    (zoomed: $x < 0.5$)")
axs[1].set_xlabel("x [rad]")
axs[1].set_ylabel("u [m/s]")

# Range III
m2 = x < 0.02
axs[2].plot(x[m2], u_dissipation[m2], color='tab:green', lw=0.8)
axs[2].set_title(fr"Range III: Dissipation Range: $k > {Kyita}$   (zoomed: $x < 0.02$)")
axs[2].set_xlabel("x [rad]")
axs[2].set_ylabel("u [m/s]")

for ax in axs:
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
